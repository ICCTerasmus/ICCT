## Gyors Fourier transzformáció (FFT)

A gyors Fourier transzformáció (FFT) egy algoritmus, ami kiszámítja a a diszkrét Fourier transzformációját (DFT) egy sorozatnak, vagy annak inverzét (IDFT). A Fourier transzformáció egy jelet az eredeti (átalában idő vagy tér) tartományából frekcencia tartományba konvertálja, illetve onnan vissza.

A példa bemutatja (vizuálisan) a jelek transzformációját idptartományból frekvencia tartományba. Három testreszabható jelkomponens segítségével létrehozható egy összetett jel (első és második ábra), aminek a rendszer Fourier transzformáció segítségével meghatározza a frekvencia tartományú reprezentációját. Ezen felül tetszőlegesen hozzáadható zaj is a jelhez.

A következő függvények használhatóak fel a jel kompozícióhoz:
* szinusz hullám,
* koszinusz hullám,
* csillapított hullám,
* rámpa függvény,
* egységugrás függvény.
