## Kifutón lévő repülőgép útvonal szabályozása

A forgási karakszerisztikája egy felszínen haladó repükőnek az alábbi módon adható meg:

$$
J_z\ddot{\psi} = bF_1\delta + F_2\dot{\psi} \, , 
$$

ahol $J_z = 11067000$ kg$\text{m}^2$, $b = 15$ , $F_1 = 35000000$ Nm, $F_2 = 500000$ kg$\text{m}^2$/$\text{s}$, $\psi$ az orientáció (radiánban), vagy az elfordulás szöge a függőleges tengelyhez mentén és $\delta$ az első kerék kormányszöge (radiánban). Amikor a repülő egy egyenes vonalat követ longitudiális irányú $V$ sebességgel (m/s-ban), akkor a repülő oldalirányú sebessége $V_y$ (m/s-ban) megközelítőleg egyenesen arányos az elfordulás szögével: $V_y = \dot{p_y} = V\psi$.

A cél egy olyan regulátor tervezése a repülő oldalirányú pozíciója $p_y$ számára, ami $V$ = 35 km/h sebesség mellett, az első kerék $\delta$ szögét kormányozásra használva az alábbi paraméterek szerint működik:
- a beállási idő 5%-os határértéken belülre kevesebb, mint 4 másodperc;
- nincs maradó hiba;
- nincs, vagy minimális a túllövés;
- a kormányzó szög a $\delta$ tartományon belül marad, ha az ugrás 5 méternél kisebb.

A rendszer állapotegyenletei:

\begin{cases}
    \dot{x} = \begin{bmatrix} \frac{F_2}{J_z} & 0 & 0 \\ 1 & 0 & 0 \\ 0 & V & 0 \end{bmatrix}x + \begin{bmatrix} \frac{bF_1}{J_z} \\ 0 \\ 0 \end{bmatrix}u \\
    y = \begin{bmatrix} 0 & 0 & 1 \end{bmatrix}x \, ,
\end{cases}

ahol $x=\begin{bmatrix} x_1 & x_2 & x_3 \end{bmatrix}^T = \begin{bmatrix} \dot{\psi} & \psi & p_y \end{bmatrix}^T$ és $u=\delta$.

A rendszer pólusai $0$, $0$ és $\frac{F_2}{J_z} \simeq 0.045$, ezáltal a rendszer instabil.

### Regulátor tervezése
#### Szabályozó tervezése

A maradó hiba eltüntetéséhez egy új állapot bevezetése szükséges:
$$
\dot{x_4} = p_y-y_d = x_3 - y_d
$$
A kiegészített rendszer ezáltal:

\begin{cases}
    \dot{x_a} = \begin{bmatrix} \frac{F_2}{J_z} & 0 & 0 & 0 \\ 1 & 0 & 0 & 0 \\ 0 & V & 0 & 0 \\ 0 & 0 & 1 & 0 \end{bmatrix}x_a + \begin{bmatrix} \frac{bF_1}{J_z} & 0 \\ 0 & 0 \\ 0 & 0 \\ 0 & -1 \end{bmatrix}\begin{bmatrix} u \\ y_d \end{bmatrix} \\
    y_a = \begin{bmatrix} 0 & 0 & 1 & 0 \\ 0 & 0 & 0 & 1 \end{bmatrix}x_a,
\end{cases}

ahol $x_a = \begin{bmatrix} x_1 & x_2 & x_3 & x_4 \end{bmatrix}^T$ és egy második kimenet kerül hozzáadásra a megfigyelhetőség megőrzéséhez. A rendszer irányítható marad $u$ bemenet szerint, ezáltal létrehozhatunk hozzá egy állapot visszacsatolást. Egy lehetséges megoldás az összes pólus $-2$-be helyezése.

#### Megfigyelő tervezése

Habár $x_3$ és $x_4$ állapotok mérhetőek és csak $x_1$ és $x_2$ becslésére van szükség, praktikusabb a teljes 4x4-es rendszerrel dolgozni és az összes állapothoz megfigyelőt tervezni $-10$ pólusokkal.

```python
# Support functions

def eigen_choice(selec,seleo):
    if selec == '0 komplex sajátérték':
        eig1c.children[0].children[0].disabled = False
        eig2c.children[1].children[0].disabled = True
        eig3c.children[0].children[0].disabled = False
        eig4c.children[0].children[0].disabled = False
        eig4c.children[1].children[0].disabled = True
        eigc = 0
    if seleo == '0 komplex sajátérték':
        eig1o.children[0].children[0].disabled = False
        eig2o.children[1].children[0].disabled = True
        eig3o.children[0].children[0].disabled = False
        eig4o.children[0].children[0].disabled = False
        eig4o.children[1].children[0].disabled = True
        eigo = 0
    if selec == '2 komplex sajátérték':
        eig1c.children[0].children[0].disabled = False
        eig2c.children[1].children[0].disabled = False
        eig3c.children[0].children[0].disabled = False
        eig4c.children[0].children[0].disabled = True
        eig4c.children[1].children[0].disabled = True
        eigc = 2
    if seleo == '2 komplex sajátérték':
        eig1o.children[0].children[0].disabled = False
        eig2o.children[1].children[0].disabled = False
        eig3o.children[0].children[0].disabled = False
        eig4o.children[0].children[0].disabled = True
        eig4o.children[1].children[0].disabled = True
        eigo = 2
    if selec == '4 komplex sajátérték':
        eig1c.children[0].children[0].disabled = True
        eig2c.children[1].children[0].disabled = False
        eig3c.children[0].children[0].disabled = True
        eig4c.children[0].children[0].disabled = False
        eig4c.children[1].children[0].disabled = False
        eigc = 4
    if seleo == '4 komplex sajátérték':
        eig1o.children[0].children[0].disabled = True
        eig2o.children[1].children[0].disabled = False
        eig3o.children[0].children[0].disabled = True
        eig4o.children[0].children[0].disabled = False
        eig4o.children[1].children[0].disabled = False
        eigo = 4
    return eigc, eigo

def method_choice(selm):
    if selm == 'K és L megadása':
        method = 1
        selec.disabled = True
        seleo.disabled = True
    if selm == 'Sajátértékek megadása':
        method = 2
        selec.disabled = False
        seleo.disabled = False
    return method
```
