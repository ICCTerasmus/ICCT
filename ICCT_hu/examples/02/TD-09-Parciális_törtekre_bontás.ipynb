## Parciális törtekre bontás - függvény bemenet

Laplace transzformációt alkalmazó rendszerelemzés esetében a kimeneti jel a bemenet és a rendszer Laplace transzformáltjainak szorzata. Ennek a szorzásnak az eredménye általában nehezen értelmezhető. Ahhoz, hogy az eredményt inverz Laplace transzformálhassuk, előtte célszerű azt parciális törtekre bontani.
A példa ezt a folyamatot mutatja be.

---

### Hogyan kezelhető a példa?
Váltogathaz a *Függvény bemenet* ls a *Polinom együttható bemenet* opciók között.

1. *Függvény bemenet*:
 * Példa: Az $\frac{1}{x^2(x^2 + 1)}$ (szorzat alak) függvény bevitelére gépelje be az 1/(x\*\*2\*(x\*\*2 + 1)) karaktersort. Az azonos függvény összevont formájában ($\frac{1}{x^4+x^2}$) a 1/(x\*\*4+x\*\*2) karakterlánccal definiálható.

2. *Polinom együttható bemenet*:
 * A csúszkákkal kiválaszthatja a számláló és nevező polinomok fokszámát.
 * A polinom együtthatók a megfelelő beviteli mezőkön át adhatók meg, és a *Megerősít* gombbal rögzíthetők.
