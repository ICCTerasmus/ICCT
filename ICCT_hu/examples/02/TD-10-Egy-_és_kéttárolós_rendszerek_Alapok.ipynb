## Egytárolós rendszerek zérusok nélkül

### Bevezetés

Elgytárolós, zérus nelküli rendszerek az alábbi átviteli függvénnyel irhatók le:

\begin{equation}
    G(s)=\frac{k}{s+k}.
\end{equation}

Ahol a $k$ érték fontos, mivel az alábbi paramétereket határozza meg:
- $1/k$ jelöli az *időállandó*-t, ami kifejezi, hogy egységugrás bemenet esetén a rendszer mennyi idő alatt éri el $\approx$ 63\%-át a végértéknek.
- $t_r$ jelöli a *felfutási idő*-t, az időtartamot, ami alatt a rendszer felfutásakor elér a végérték 10\%-ról a 90\%-ára.
- $t_s$ jelöli a *beállási időt*, az időtartamot, mialatt a rendszer végleg beáll a hibahatáron belülre (pl. 2\%, mint a példában).

Az ugrásválasz az alábbiként értelmezett:

\begin{equation}
    c(t)=1-e^{-at},
\end{equation}

ahol a gerjesztett válasz $1$ és a gerjesztetlen válasz $-e^{-at}$.

## Kéttárolós rendszerek

### Bevezetés

Az egytárolós rendszerekkel szemben, ahol a $k$ paraméter csak a válasz sebességét befolyásolta, a kéttárolós rendszerek hasonló paramétereinek változtatásával a kimenet alakja is változhat. A következő négy változata lehetséges a kimeneti jelnek:
- *túlcsillapított* válasz,
- *alulcsillapított* válasz,
- *csillapítatlan* válasz, és
- *kritikusan csillapított* válasz.

### Hogyan kezelhető a példa?

A csúszkák segítségével megválasztható az $a$ és $b$ paramétere a kéttárolós rendszer $G(s)=\frac{b}{s^2+as+b}$ átviteli függvényének és megfigyelhető a rendszer pólus-zérus görbéje, valamint ugrásválasza. 

```python
# set up plot
fig1, ax1 = plt.subplots(1,2,figsize=[9.8,4],num='Kéttárolós rendszer')
ax1[0].set_ylim([-3.5, 3])
ax1[1].set_ylim([0, 2.5])
# ax1.set_xlim([0, 5])
ax1[0].grid(True)
ax1[1].grid(True)
ax1[0].axhline(y=0,color='k',lw=.8)
ax1[1].axhline(y=0,color='k',lw=.8)
ax1[0].axvline(x=0,color='k',lw=.8)
ax1[1].axvline(x=0,color='k',lw=.8)
ax1[0].set_xlabel('Re')
ax1[0].set_ylabel('Im')
ax1[1].set_xlabel('$t$ [s]')
ax1[1].set_ylabel('Bemenet, kimenet')
ax1[0].set_title('Pólus-zérus görbe')
ax1[1].set_title('Időfüggvény')

t = np.linspace(0, 20, 10000)

textGs = ax1[0].text(0,0,'')

ax1[1].step([0,20],[0,1],color='C0',label='bemenet')

plotzero, = ax1[0].plot([], [])
response2, = ax1[1].plot([], [])

def response_func2(t, a, b):
    num_sys=np.array([b])
    den_sys=np.array([1,a,b])
    tf_sys=c.TransferFunction(num_sys,den_sys)
    poles_sys,zeros_sys=c.pzmap(tf_sys, Plot=False)
    T, yout = c.step_response(tf_sys,t)
    return T, yout, poles_sys, tf_sys
 
@widgets.interact(a=(0, 10, 1),b=(1,10,1))

def update(a=7,b=9):
    """ Update plots """
    global response2, plotzero, textGs
    ax1[0].lines.remove(plotzero)
    ax1[1].lines.remove(response2)
#     textGs.remove()
    T, yout, poles_sys, tf_sys = response_func2(t, a, b)
    plotzero, = ax1[0].plot(np.real(poles_sys), np.imag(poles_sys), 'xg', markersize=10, label = 'Pólusok')
#     textGs = ax1[0].text(-7,1,tf_sys)
    response2, = ax1[1].plot(T,yout,color='C1',label='kimenet')
    s=sym.Symbol('s')
    eq=b/(s**2+a*s+b)
    coeff = [1,a,b]
    rootsdenom=np.roots(coeff)
    eq2=b/((s-rootsdenom[0])*(s-rootsdenom[1]))
    display(Markdown('A rendszer átviteli függvénye $G(s)$:'))
    display(eq),display(Markdown('vagy')),display(eq2)

    if np.imag(poles_sys)[0] == 0 and np.imag(poles_sys)[1] == 0 and np.real(poles_sys)[0] < 0 and np.real(poles_sys)[1] < 0 and np.real(poles_sys)[0]!=np.real(poles_sys)[1]:
        display(Markdown('A rendszer **túlcsillapított**, mert mindkét pólus negatív és valós.'))
    elif math.isclose(0, np.imag(poles_sys)[0], abs_tol=10**-6) and math.isclose(0, np.imag(poles_sys)[1], abs_tol=10**-6) and np.real(poles_sys)[1] < 0 and np.real(poles_sys)[0]==np.real(poles_sys)[1]:
        display(Markdown('A rendszer **kritikusan csillapított**, mert a komplex konjugált póluspár képzeres része nulla.'))
    elif np.real(poles_sys)[0] == 0 and np.real(poles_sys)[1] == 0:
        display(Markdown('A rendszer **csillapítatlan**, mert a komplex konjugált póluspár valós része nulla.'))
    elif np.imag(poles_sys)[0] != 0 and np.imag(poles_sys)[1] != 0 and np.real(poles_sys)[0] != 0 and np.real(poles_sys)[1] != 0:
        display(Markdown('A rendszer **alulcsillapított**, mert a komplex konjugált póluspár mindkét tagja nem-nulla.'))
    ax1[0].legend()
    ax1[1].legend()
```
