## Linearizacija funkcij

### Uvod

> Linearizacija je definirana kot proces določitve linearne aproksimacije izbrane funkcije okoli izbrane točke. Linearno aproksimacijo funkcije dobimo z razvojem Taylorjeve vrste okoli izbrane točke, pri čemer ohranimo zgolj prva dva člena. Linearizacija je učinkovita metoda za aproksimacijo funkcije $y=f(x)$ pri katerikoli vrednosti $x=x_0+\Delta x$ na podlagi vrednosti in naklona funkcije pri $x=x_0+\Delta x$ (ob predpostavki, da je funkcija $f(x)$ odvedljiva na intervalu $[x_0,x_0+\Delta x]$ (ali $[x_0+\Delta x,x_0]$) in da je $x_0$ blizu $x_0+\Delta x$. Na kratko, linearizacija aproksimira vrednost funkcije v bližini točke $x=x_0$. (vir: [Wikipedia](https://en.wikipedia.org/wiki/Linearization))

V tem interaktivnem primeru je linearizacija definirana kot:

\begin{equation}
    f(x)\approx f(x_0)+f^{\prime}(x_0) \cdot (x-x_0),
\end{equation}

kjer je $f^{\prime}=\frac{f(x_0+h)-f(x_0)}{h}$ (vrednost $h$ je nastavljena na $0.01$).

Enotska skočna funkcija je definirana kot: 

\begin{equation}
    u(x) =
    \begin{cases}
      0; & \text{$x<0$}\\
      1; & \text{$x\geq0$}
    \end{cases},
\end{equation}

in enotska rampa kot: 

\begin{equation}
    r(x) =
    \begin{cases}
      0; & \text{$x<0$}\\
      x; & \text{$x\geq0$}      
    \end{cases}.
\end{equation}   
    
---

### Kako upravljati s tem interaktivnim primerom?
Z uporabo drsnikov spreminjaj vrednost $x_0$, tj.  vrednosti $x$ pri kateri želiš linearizirati funkcijo.

```python
# sinus, step, ramp, x^2, sqrt(x)
style = {'description_width': 'initial','button_width':'180px'}
functionSelect = widgets.ToggleButtons(
    options=[('sinusna funkcija', 0), ('enotska skočna funkcija', 1), ('enotska rampa', 2), ('parabolična funkcija', 3), ('korenska funkcija', 4)],
    description='Izberi: ',
    style=style)

fig = plt.figure(num='Linearizacija funkcije')
fig.set_size_inches((9.8, 3))
fig.set_tight_layout(True)
f1 = fig.add_subplot(1, 1, 1)

f1.grid(which='both', axis='both', color='lightgray')

f1.set_xlabel('$x$')
f1.set_ylabel('$f(x)$')

f1.axhline(0,Color='black',linewidth=0.5)
f1.axvline(0,Color='black',linewidth=0.5)

func_plot, = f1.plot([],[])
tang_plot, = f1.plot([],[])
point_plot, = f1.plot([],[])

f1.set_xlim((-5,5))
f1.set_ylim((-6,6))

def create_draw_functions(x0,index):
    x=np.linspace(-5,5,1001)
    h=0.001 # equal to \Delta x
    
    global func_plot, tang_plot, point_plot
       
    if index==0:
        y=np.sin(x)
        fprime=(np.sin(x0+h)-np.sin(x0))/h
        tang=np.sin(x0)+fprime*(x-x0)
        fx0=np.sin(x0)        
    elif index==1:
        y=np.zeros(1001)
        y[510:1001]=1
    elif index==2:
        y=np.zeros(1001)
        y[500:1001]=np.linspace(0,5,501)
    elif index==3:
        y=x*x
        fprime=((x0+h)*(x0+h)-(x0*x0))/h
        tang=x0*x0+fprime*(x-x0)
        fx0=x0*x0        
    elif index==4:
        x1=np.linspace(0,5,500)
        y=np.sqrt(x1)
        if x0>=0:
            fprime=(np.sqrt(x0+h)-np.sqrt(x0))/h
            tang=np.sqrt(x0)+fprime*(x-x0)
            fx0=np.sqrt(x0)
    
    f1.lines.remove(func_plot)
    f1.lines.remove(tang_plot)
    f1.lines.remove(point_plot)
    
    if index == 0:
        func_plot, = f1.plot(x,y,label='$f(x)=sin(x)$',color='C0')
        tang_plot, = f1.plot(x,tang,'--r',label='tangenta')
        point_plot, = f1.plot(x0,fx0,'om',label='$x_0$')
        for txt in f1.texts: 
                txt.set_visible(False)
    elif index == 1: # in case of the unit step function
        if x0==0:
            func_plot, = f1.step(x,y,label='$f(x)=u(x)$',color='C0')
            tang_plot, = f1.plot([],[])
            point_plot, = f1.plot([],[])    
            f1.text(0.1,1.3,'Linearizacija pri $x_0=0$ ni možna!',fontsize=14)
        elif x0<0:
            tang=np.zeros(1001)
            func_plot, = f1.step(x,y,label='$f(x)=u(x)$',color='C0')
            tang_plot, = f1.plot(x,tang,'--r',label='tangenta')
            point_plot, = f1.plot(x0,[0],'om',label='$x_0$')
            for txt in f1.texts: 
                txt.set_visible(False)
        elif x0>0:
            tang=np.ones(1001)
            func_plot, = f1.step(x,y,label='$f(x)=u(x)$',color='C0')
            tang_plot, = f1.plot(x,tang,'--r',label='tangenta')
            point_plot, = f1.plot(x0,[1],'om',label='$x_0$')
            for txt in f1.texts: 
                txt.set_visible(False)
    elif index==2: # in case of the ramp
        if x0<0:
            tang=np.zeros(1001)
            func_plot, = f1.plot(x,y,label='$f(x)=R(x)$',color='C0')
            tang_plot, = f1.plot(x,np.zeros(1001),'--r',label='tangenta')
            point_plot, = f1.plot(x0,[0],'om',label='$x_0$')
            for txt in f1.texts: 
                txt.set_visible(False)
        elif x0>=0:
            tang=x
            func_plot, = f1.plot(x,y,label='$f(x)=R(x)$',color='C0')
            tang_plot, = f1.plot(x,tang,'--r',label='tangenta')
            point_plot, = f1.plot(x0,x0,'om',label='$x_0$')
            for txt in f1.texts: 
                txt.set_visible(False)
    elif index==3:
        func_plot, = f1.plot(x,y,label='$f(x)=x^2$',color='C0')
        tang_plot, = f1.plot(x,tang,'--r',label='tangenta')
        point_plot, = f1.plot(x0,fx0,'om',label='$x_0$')
        for txt in f1.texts: 
                txt.set_visible(False)        
    elif index==4: #in case of the square root function
        if x0<0:
            for txt in f1.texts: 
                txt.set_visible(False)
            func_plot, = f1.plot(x1,y,label='$f(x)=\sqrt{x}$',color='C0')
            tang_plot, = f1.plot([],[])
            point_plot, = f1.plot([],[])
            f1.text(-4.9,1.3,'Korenska funkcija ni definirana za vrednosti $x<0$!',fontsize=14)
        else:
            func_plot, = f1.plot(x1,y,label='$f(x)=\sqrt{x}$',color='C0')
            tang_plot, = f1.plot(x,tang,'--r',label='tangent')
            point_plot, = f1.plot(x0,fx0,'om',label='$x_0$')
            for txt in f1.texts: 
                txt.set_visible(False)
    
    if (index==1) and x0==0 or (index==4 and x0<0):
        display(Markdown('Preberi komentar na grafu.'))
    else:
        k=round(((tang[-1]-tang[0])/(x[-1]-x[0])),3)
        n=round(((tang[-1]-k*x[-1])),3)
        display(Markdown('Enačba tangente: $y=%.3fx+%.3f$.'%(k,n)))
    
    f1.legend()
    
    f1.relim()
    f1.relim()
    f1.autoscale_view()
    f1.autoscale_view()   

    
x0_slider = widgets.FloatSlider(value=1, min=-5, max=5, step=0.2, description='$x_0$',
                          continuous_update=True, layout=widgets.Layout(width='auto', flex='5 5 auto'),readout_format='.1f')

input_data = widgets.interactive_output(create_draw_functions, {'x0':x0_slider, 'index':functionSelect})

def update_sliders(index):
    global x0_slider
    
    x0val = [0.5, 0.5, 1, 1, 5, 10]
    x0slider.value = x0val[index]
    
input_data2 = widgets.interactive_output(update_sliders, {'index':functionSelect})

display(functionSelect)

display(x0_slider,input_data)

# display(Markdown("The system can be represented as $f(x)=5$ for small excursions of x about x0."))
```
