## Spoznavnost

Za dan linearen sistem

\begin{cases}
\dot{\textbf{x}}=A\textbf{x}+B\textbf{u} \\ 
\textbf{y}=C\textbf{x}
\end{cases}

želimo vedeti, ali lahko, na podlagi časovnega poteka izhoda $y(t)$ in vhoda $u(t)$, določimo začetno stanje sistema, ki je povzročilo dani odziv (tj. izhod). Tovrstni tip informacije o sistemu je povezan z lastnostjo sistema, ki jo imenujemo **spoznavnost** (ang. *observability*).

Najprej uvedimo koncept <u>nerazločljivosti</u> med dvema stanjema. 
>Dve stanji $\bar{x_1}$ in $\hat{x_1}$ sta nerazločljivi v času ($\bar{x_1}I_{t_2}\hat{x_1}$), _če in samo če_ $\forall u \in U \quad y(t,t_1,\bar{x_1},u)=y(t,t_1,\hat{x_1},u) \quad \forall t\in [t_1,t_2]$, kjer je $U$ set vhodnih vrednosti. 

Z drugimi besedami: dve stanji sta nerazločljivi v času, če in samo če ob enakem vhodu $u(t)$ dosežemo enak izhod $y(t)$ v izbranem časovnem intervalu. Pokažemo lahko, da sta sicer za zvezne, linearne časovno nespremenljive sisteme, nerazločljivost in spoznavnost neodvisni od izbranega časovnega intervala.

Velja omeniti, da je nerazločljivost med dvema stanjema $x_1$ and $x_2$ enaka nerazločljivosti med razliko med obema stanjema in izhodiščem prostora stanj, tj. med $x_1-x_2$ in $0$. Enačitev izhodnih odzivov iz dveh različnih začetnih pogojev lahko zapišemo kot:

\begin{cases}
\bar{y}(t) = Ce^{At}\bar{x_1} + C\int_{t_1}^{t}e^{A(t-\tau)}Bu(\tau)d\tau \\ 
\hat{y}(t) = Ce^{At}\hat{x_1} + C\int_{t_1}^{t}e^{A(t-\tau)}Bu(\tau)d\tau
\end{cases}

in dobimo:

$$
Ce^{At}\bar{x_1} = Ce^{At}\hat{x_1} \rightarrow Ce^{At}\underbrace{(\bar{x_1}-\hat{x_1})}_{\widetilde{x_1}} = Ce^{At}0 \,.
$$

Na podlagi zgornjega izraza lahko govorimo 1) o nerazločjivosti glede na izhodišče prostora stanj in 2) o tem, da je stanje nespoznavno, če je nerazločljivo glede na izhodišče.

Merilo spoznavnosti sistema lahko uvedemo z uporabo spoznavnostne Gramove matrike 
$G_o(0,t_f)=\int_{0}^{t_f}e^{A^T\tau}C^TCe^{A\tau}d\tau$. Pokažemo lahko, da je prostor nespoznavnih stanj sistema enak jedru $G_o$; linearni sistem je spoznaven, če in samo če $\text{ker}(G_o)=\{0\}$, oz. (ekvivalentno) če velja $\widetilde{x}I_{t}0 \Rightarrow \widetilde{x}=\{0\}$.

Če definiramo spoznavno matriko kot

$$
\mathcal{O} = \begin{bmatrix}C \\ CA \\ CA^2 \\ \vdots \\ CA^{n-1} \end{bmatrix}
$$

lahko pokažemo, da $\text{ker}(G_o)=\text{ker}(\mathcal{O})$; spoznavnost sistema lahko tako določimo glede na rang spoznavnostne matrike, tj. če $\text{rank}(\mathcal{O})=n$, kjer je $n: \, x\in \mathbb{R}^n$, potem je sistem spoznaven. 

Znanje o spoznavnosti sistema in vsakega posameznega stanja je ključno, saj vpliva na to, katere modalne oblike bodo prisotne na izhodu in katere ne. Če je sistem spoznaven, se lahko vse njegove modalne oblike pojavijo na izhodu (odvisno od začetnih pogojev in vhodnih funkcij). Če sistem ni spoznaven, tj. če $\mathcal{O}$ ni polni rang, potem nekatere modalne oblike sistema niso vidne na izhodu.

**Opomba:** Koncept analogen spoznavnosti je obnovljivost (ang. *reconstructability*), ki temelji na obnovi/rekonstrukciji končnega stanja, v katerem se nahaja sistem, namesto začetnega stanja. Za časovne, zvezne linearne sisteme je sistem obnovljiv, če in samo če je spoznaven, kar pa ne velja za diskretne sisteme. To lahko preprosto dokažemo s tem, da kadar je začetno stanje spoznavno, potem lahko trenutno stanje obnovimo z uporabo Lagrangeovih enačb (ker je vhod $u$ znan). Če pa je znano trenutno stanje, lahko začetno stanje dobimo z integracijo po času.


### Kako upravljati s tem interaktivnim primerom?

V tem interaktivnem primeru lahko spreminjaš vrednosti elementov matrik $A$ in $C$:
- Poizkusi doseči, da bo nespoznavno zgolj tretje stanje sistema $x_1$ para $(A,C)$, le s spreminjanjem vrednosti elementov matrike $C$.
- Poizkusi doseči, da bo nespoznavno zgolj drugo stanje sistema $x_2$ para $(A,C)$, le s spreminjanjem vrednosti elementov matrike $C$.
- Poizkusi doseči, da bo par $(A,C)$ nespoznaven na način, da spreminjaš zgolj vrednosti elementov matrike $C$. Ko to dosežeš, pomisli, zakaj je matrika $A$ postala nespoznavna. Za katera stanja se je spoznavnost izgubila?
- Za dano matriko $C=[1, 1, 1]$ določi matriko $A$ tako, da bo par $(A,C)$ nespoznaven (pomagaj si z uporabo matrike $\mathcal{O}$).
- Poizkusi zapisati par $(A,C)$ tako, da bo ena pozitivna vrednost povezana z nespoznavnim stanjem. Kakšen je prosti odziv sistema?

```python
#Preparatory Cell 

%matplotlib inline
import control
import numpy
from IPython.display import display, Markdown
import ipywidgets as widgets
import matplotlib.pyplot as plt

#print a matrix latex-like
def bmatrix(a):
     """Returns a LaTeX bmatrix - by Damir Arbula (ICCT project)

     :a: numpy array
     :returns: LaTeX bmatrix as a string
     """
     if len(a.shape) > 2:
         raise ValueError('bmatrix can at most display two dimensions')
     lines = str(a).replace('[', '').replace(']', '').splitlines()
     rv = [r'$\begin{bmatrix}']
     rv += ['  ' + ' & '.join(l.split()) + r'\\' for l in lines]
     rv +=  [r'\end{bmatrix}$']
     return ''.join(rv) #'\n'.join(rv)

#matrixWidget is a matrix looking widget built with a VBox of HBox(es) that returns a numPy array as value !
class matrixWidget(widgets.VBox):
    def updateM(self,change):
        for irow in range(0,self.n):
            for icol in range(0,self.m):
                self.M_[irow,icol] = self.children[irow].children[icol].value
                #print(self.M_[irow,icol])
        self.value = self.M_

    def dummychangecallback(self,change):
        pass
    
    
    def __init__(self,n,m):
        self.n = n
        self.m = m
        self.M_ = numpy.matrix(numpy.zeros((self.n,self.m)))
        self.value = self.M_
        widgets.VBox.__init__(self,
                             children = [
                                 widgets.HBox(children = 
                                              [widgets.FloatText(value=0.0, layout=widgets.Layout(width='90px')) for i in range(m)]
                                             ) 
                                 for j in range(n)
                             ])
        
        #fill in widgets and tell interact to call updateM each time a children changes value
        for irow in range(0,self.n):
            for icol in range(0,self.m):
                self.children[irow].children[icol].value = self.M_[irow,icol]
                self.children[irow].children[icol].observe(self.updateM, names='value')
        #value = Unicode('example@example.com', help="The email value.").tag(sync=True)
        self.observe(self.updateM, names='value', type= 'All')
        
    def setM(self, newM):
        #disable callbacks, change values, and reenable
        self.unobserve(self.updateM, names='value', type= 'All')
        for irow in range(0,self.n):
            for icol in range(0,self.m):
                self.children[irow].children[icol].unobserve(self.updateM, names='value')
        self.M_ = newM
        self.value = self.M_
        for irow in range(0,self.n):
            for icol in range(0,self.m):
                self.children[irow].children[icol].value = self.M_[irow,icol]
        for irow in range(0,self.n):
            for icol in range(0,self.m):
                self.children[irow].children[icol].observe(self.updateM, names='value')
        self.observe(self.updateM, names='value', type= 'All')        

                #self.children[irow].children[icol].observe(self.updateM, names='value')

             
#overlaod class for state space systems that DO NOT remove "useless" states (what "professor" of automatic control would do this?)
class sss(control.StateSpace):
    def __init__(self,*args):
        #call base class init constructor
        control.StateSpace.__init__(self,*args)
    #disable function below in base class
    def _remove_useless_states(self):
        pass
```

```python
# Preparatory cell
A = numpy.matrix('-1 1 0; 0 -1 1; 0 0 -1')
B = numpy.matrix('0; 0; 1')
C = numpy.matrix('1 0 0')
X0 = numpy.matrix('2; 2; 2')

Aw = matrixWidget(3,3)
Aw.setM(A)
Bw = matrixWidget(3,1)
Bw.setM(B)
Cw = matrixWidget(1,3)
Cw.setM(C)
X0w = matrixWidget(3,1)
X0w.setM(X0)
```

```python
# Misc
#define type of ipout 
selu = widgets.Dropdown(
    options=['impulzna funkcija', 'koračna funkcija', 'sinusoidna funkcija', 'kvadratni val'],
    value='koračna funkcija',
    description='Vhod:',
    disabled=False
)
# Define the values of the input
u = widgets.FloatSlider(
    value=0,
    min=0,
    max=20.0,
    step=0.1,
    description=r'$u$:',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='.1f',
)
period = widgets.FloatSlider(
    value=1,
    min=0.01,
    max=4,
    step=0.01,
    description='Perioda: ',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='.2f',
)

#create dummy widget 
DW = widgets.FloatText(layout=widgets.Layout(width='0px', height='0px'))

#create button widget
START = widgets.Button(
    description='Test',
    disabled=False,
    button_style='', # 'success', 'info', 'warning', 'danger' or ''
    tooltip='Test',
    icon='check'
)
                       
def on_start_button_clicked(b):
    #This is a workaround to have intreactive_output call the callback:
    #   force the value of the dummy widget to change
    if DW.value> 0 :
        DW.value = -1
    else: 
        DW.value = 1
    pass
START.on_click(on_start_button_clicked)
```

```python
def main_callback(A, B, C, X0, u, period, selu, DW):
    eig = numpy.linalg.eig(A)[0]
    O = control.obsv(A,C)
    rankO = numpy.linalg.matrix_rank(O)
    
    text = r'Lastne vrednosti sistema so ' + str(eig[0]) + r', ' + str(eig[1])+ r' in ' + str(eig[2]) + r'. '
    if rankO == 3:
        text = text + r'Ker $\text{rank}(\mathcal{O})=3$, je sistem spoznaven. ' + '\n\n' + 'Spoznavnostna matrika je: ' + r'$\mathcal{O}=$' + bmatrix(O) #+ r'$'
    else:
        text = text + r'Ker $\text{rank}(\mathcal{O})=' + str(rankO) + r'$, je sistem nespoznaven. ' + '\n\n' + 'Spoznavnostna matrika je: ' + r'$\mathcal{O}=' + bmatrix(O) + r'$'
    display(Markdown(text))
    
    
    sys = sss(A,B,C,0)
    if numpy.real([eig[0],eig[1],eig[2]])[0] == 0 and numpy.real([eig[0],eig[1],eig[2]])[1] == 0 and numpy.real([eig[0],eig[1],eig[2]])[2] == 0:
        T = numpy.linspace(0,20,1000)
    else:
        if min(numpy.abs(numpy.real([eig[0],eig[1],eig[2]]))) != 0:
            T = numpy.linspace(0,7*1/min(numpy.abs(numpy.real([eig[0],eig[1],eig[2]]))),1000)
        else:
            T = numpy.linspace(0,7*1/max(numpy.abs(numpy.real([eig[0],eig[1],eig[2]]))),1000)
    
    if selu == 'impulzna funkcija': #selu
        U = [0 for t in range(0,len(T))]
        U[0] = u
        T, Y, X = control.forced_response(sys,T=T,U=U,X0=X0)
    if selu == 'koračna funkcija':
        U = [u for t in range(0,len(T))]
        T, Y, X = control.forced_response(sys,T=T,U=U,X0=X0)
    if selu == 'sinusoidna funkcija':
        U = u*numpy.sin(2*numpy.pi/period*T)
        T, Y, X = control.forced_response(sys,T=T,U=U,X0=X0)
    if selu == 'kvadratni val':
        U = u*numpy.sign(numpy.sin(2*numpy.pi/period*T))
        T, Y, X = control.forced_response(sys,T=T,U=U,X0=X0)
    
    fig = plt.figure(figsize=(16,5))
    
    fig.add_subplot(121)
    plt.title('Začetni odziv: stanja')
    plt.plot(T,X[0])
    plt.plot(T,X[1])
    plt.plot(T,X[2])
    plt.ylabel('Stanja')
    plt.xlabel('čas [s]')
    plt.legend([r'$x_1$',r'$x_2$',r'$x_3$'])
    plt.grid()
    
    fig.add_subplot(122)
    plt.ylabel(r'$y$')
    plt.plot(T,Y)
    plt.xlabel('čas [s]')
    plt.title('Začetni odziv: izhod')
    plt.grid()
    
    
alltogether = widgets.VBox([widgets.HBox([widgets.Label('A:',border=3),  Aw, widgets.Label('   ',border=3),
                                          widgets.Label('B:',border=3),  Bw, widgets.Label('   ',border=3),
                                          widgets.Label('x0:',border=3),X0w,widgets.Label('   ',border=3),
                                          START]),
                            widgets.Label('   ',border=3),
                            widgets.HBox([widgets.Label('C:',border=3),Cw,widgets.Label('   ',border=3)]),
                            widgets.Label('   ',border=3),
                            widgets.HBox([selu, u, period])])
out = widgets.interactive_output(main_callback, {'A':Aw, 'B':Bw, 'C':Cw, 'X0':X0w, 'u':u, 'period':period, 'selu':selu, 'DW':DW})
out.layout.height = '460px'
display(out, alltogether)
```
